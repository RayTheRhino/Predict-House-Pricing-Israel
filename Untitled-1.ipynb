```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
import math
```

```python
data = pd.read_csv("Nadlan_info_TLV.csv")
```

```python
data
```

```python
data['worth'] = data['worth'].str.replace(',', '').astype(float)
data['change'] = data['change'].str.replace(',', '').astype(float)
```

```python
data.info
```

```python
data.dropna(inplace=True)
data.drop(['bloock-smooth'] ,axis=1,inplace=True)
data.drop(['essence'] ,axis=1,inplace=True)
data.drop(['date of sale'] ,axis=1,inplace=True)
data.drop(['change'],axis=1,inplace=True)
```

```python
data.info
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
x = data.drop(['worth'] ,axis=1)
y = data['worth']
```

```python
x
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="worth", y="year built", data=data, hue="worth", palette="coolwarm")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="worth", y="Area", data=data, hue="worth", palette="coolwarm")
```

```python
scaler = StandardScaler()
columns_to_scale = ['Area', 'year built','part sold','Room number']
x[columns_to_scale] = scaler.fit_transform(x[columns_to_scale])
x
```

```python
dummy_df = pd.get_dummies(x['settelment'], prefix='settelment')

# Concatenate the dummy variables with the original DataFrame
x = pd.concat([x, dummy_df], axis=1)
x.drop(['settelment'],axis=1,inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train = x_train.astype('float32')
y_train = y_train.astype('float32')
x_test = x_test.astype('float32')
y_test = y_test.astype('float32')
```

```python
from sklearn.linear_model import SGDRegressor

reg = SGDRegressor()

reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
error = math.sqrt(mse)
print("Error: ",error)

# Calculate R-squared score
r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```
